```python
import torch
import torch.nn as nn

class CrossAttention(nn.Module):
  def __init__(self, hidden_dim):
    super().__init__()
    self.query_projection = nn.Linear(hidden_dim, hidden_dim)
    self.key_projection = nn.Linear(hidden_dim, hidden_dim)
    self.value_projection = nn.Linear(hidden_dim, hidden_dim)
    self.softmax = nn.Softmax(dim=-1)

  def forward(self, query, key, value):
    query = self.query_projection(query)
    key = self.key_projection(key)
    value = self.value_projection(value)
    attention_scores = torch.matmul(query, key.transpose(-2, -1)) / (key.size(-1) ** 0.5)
    attention_weights = self.softmax(attention_scores)
    attended_values = torch.matmul(attention_weights, value)
    return attended_values
```

```python
from transformers import BertModel

# class CrossModalModel(nn.Module):
#     def __init__(self, text_model_name='bert-base-uncased', vision_model_name='google/vit-base-patch16-224', hidden_dim=768, num_cross_attention_layers=1):
#       super().__init__()
#       self.text_encoder = BertModel.from_pretrained(text_model_name)
#       self.vision_encoder = ViTModel.from_pretrained(vision_model_name)
#       self.cross_attention_layers = nn.ModuleList(
#           [CrossAttention(hidden_dim=hidden_dim) for _ in range(num_cross_attention_layers)]
#       )
#       self.output_layer = nn.Linear(hidden_dim, 2)

#     def forward(self, text_inputs, image_features):
#       text_outputs = self.text_encoder(**text_inputs).last_hidden_state
#       image_outputs = self.vision_encoder(pixel_values=image_inputs).last_hidden_state
#       fused_outputs = text_outputs
#       for cross_attention in self.cross_attention_layers:
#           fused_outputs = cross_attention(fused_outputs, image_outputs, image_outputs)
#       task_output = self.output_layer(fused_outputs[:, 0])

#       return fused_outputs, task_output
class CrossModalModel(nn.Module):
    def __init__(self, text_model_name='bert-base-uncased', vision_model_name='google/vit-base-patch16-224', hidden_dim=768, num_cross_attention_layers=1):
        super().__init__()

        # Text encoder (BERT)
        self.text_encoder = BertModel.from_pretrained(text_model_name)

        # Vision encoder (ViT)
        self.vision_encoder = ViTModel.from_pretrained(vision_model_name)

        # Cross-attention layers
        self.cross_attention_layers = nn.ModuleList(
            [CrossAttention(hidden_dim=hidden_dim) for _ in range(num_cross_attention_layers)]
        )

        # Output layer (e.g., classification head)
        self.output_layer = nn.Linear(hidden_dim, 2)  # Example for binary classification

    def forward(self, text_inputs=None, image_inputs=None):
        """
        Args:
            text_inputs (dict): Tokenized text inputs (input_ids, attention_mask).
            image_inputs (torch.Tensor): Image features from raw images or precomputed features.

        Returns:
            torch.Tensor: Fused outputs after cross-attention.
            torch.Tensor: Final task-specific output (e.g., logits).
        """

        # Encode text inputs using BERT
        text_outputs = self.text_encoder(**text_inputs).last_hidden_state  # Shape: [batch_size, seq_len, hidden_dim]

        # Encode image inputs using ViT
        image_outputs = self.vision_encoder(pixel_values=image_inputs).last_hidden_state  # Shape: [batch_size, num_patches, hidden_dim]

        # Apply cross-attention layers
        fused_outputs = text_outputs
        for cross_attention in self.cross_attention_layers:
            fused_outputs = cross_attention(fused_outputs, image_outputs, image_outputs)  # Query is text; Key/Value is image

        # Task-specific output (e.g., classification logits)
        task_output = self.output_layer(fused_outputs[:, 0])  # Use [CLS] token embedding for classification

        return fused_outputs, task_output
```
